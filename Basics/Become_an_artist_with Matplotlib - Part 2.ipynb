# Mission 1

Create a scatter plot between any two variables x and y of your choice.

Add a caption, color it in any other color than black.


```
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0,3*np.pi,100);
y = np.cos(x) + np.random.randint(1,100);  
colors = np.random.randint(1,100,100)
plt.scatter(x, y, c=colors, cmap='hsv')
plt.title("Funny Graph", color ='red')
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Initialisierung des Zufallszahlengenerators 
rng = np.random.RandomState(0)

# Erstellung von 100 Zufallszahlen, basierend auf der "Standard-Normal"-Verteilung.
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)

# Zufällige Erstellung der Größe der einzelenen Punkte. 
sizes = 1000 * rng.rand(100)

# Plotten des Scatterplots, mit: x&y Koordinaten, sizes = Größe der Datenpunkte, aplha= Transpazenz der Datenpunkte und der Farbpalette 'viridis'.  
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')

# Erstellung der Farbskala 
plt.colorbar();  
```
