- The labels are on the right side instead of the left.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

dt = 0.01
t = np.arange(0, 30, dt)
nse1 = np.random.randn(len(t))               
nse2 = np.random.randn(len(t))               

s1 = np.sin(2 * np.pi * 10 * t) + nse1
s2 = np.sin(2 * np.pi * 10 * t) + nse2

fig, axs = plt.subplots(2, 1)
axs[0].plot(t, s1)
axs[0].plot(t, s2)
axs[0].set_xlim(0, 2)
axs[1].set_xlim(0, 2)
axs[0].set_xlabel('time')
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)

cxy, f = axs[0].cohere(s1, s2, 256, 1. / dt)
axs[0].set_ylabel('coherence')
axs[0].yaxis.set_label_position("right")
axs[0].yaxis.tick_left()

fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10) # Definition x von 0 bis 10
with plt.style.context('Solarize_Light2'): # Interessanter Wiki Artikel: https://en.wikipedia.org/wiki/Solarized_(color_scheme)

    #Plotten des Sinus von X + X multipliziert mit einer Steigung + einer Zufallszahl. Ergebnis: 8 zufällige Linien.
    # Durch das Plotten der einzelnen Linien werden die 8 Linien in den 8 unterschiedlichen Farben dargstellt.
    # Wir erkennen, dass durch Solarize_Light2 die Farben nicht "blenden". 
    plt.plot(x, np.sin(x) + x + np.random.randn(50))
    plt.plot(x, np.sin(x) + 2 * x + np.random.randn(50))
    plt.plot(x, np.sin(x) + 3 * x + np.random.randn(50))
    plt.plot(x, np.sin(x) + 4 + np.random.randn(50)) # Achtung, hier fehlt das x, hier bekommen wir also nur den sin(x) + 4 und eine Zufallszahl. 
    plt.plot(x, np.sin(x) + 5 * x + np.random.randn(50))
    plt.plot(x, np.sin(x) + 6 * x + np.random.randn(50))
    plt.plot(x, np.sin(x) + 7 * x + np.random.randn(50))
    plt.plot(x, np.sin(x) + 8 * x + np.random.randn(50))
    # Titel setzen
    plt.title('8 Random Lines - Line')
    # X,Y Achse beschriften 
    plt.xlabel('x label', fontsize=14)
    plt.ylabel('y label', fontsize=14)

plt.show()
```
