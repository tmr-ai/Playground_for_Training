# Mission 1

Create a function that takes a list as a parameter and displays only the elements (and their position) that have positions that are even (0, 2, 4, 6...).

Example list = ["a", "b", "c", "d", "e"], the function should return : 
- a position 0
- c position 2
- e position 4

```python
def even_postitions(list): 
 '''
 this function takes a list and only returns list elements that habe even positions

 list: Input List with nElements
 '''    
 for i in range(0,len(list)):
    #The % sign checks for a remainder
    if i % 2 == 0:
     print(str(list[i]) + " position " + str(i))
    
list = ["a","b","c","d","e","f"]
even_postitions(list) 
```

# Mission 4

Create a function that does the same thing as the insert method, but at a random position.

```python
from random import randint

def random_insert(list,element):
 '''
 This function adds an element to a random position in a list
 
 list: Input list 
 element: element that is beeing randomly added 
 '''
 list.insert((randint(0, len(list)+1)),str(element))
 return list 

list = [5, 5, "a", "b"]

random_insert(list, 5)
```

# Mission 5

Create a function that does the same thing as the pop method for a list containing only characters. The difference is that this function displays a message indicating the number of characters that are identical to the deleted character and still remain in the string. In other words, if there were 3 "b" s in the list, and one has just been deleted by your function, the function will display "there are still 2 letter b s".

```python
def pop_with_elements_left(list, charcater:str): 
  '''
  This function deletes a character and displays how many identicla characters are left within the list

  list: Input list 
  character: character(string) that is beeing deleted 
  '''
  if charcater not in list: 
    print("character ist not in List, therefore it can be deleted") 
  
  else:
   list.remove(charcater)
   counter = list.count(charcater)
   print(charcater + " has been deleted and ist still: " + str(counter) + " times present in list")
   return list
   
  
list = ['a', 'a', 'b' , 'B' , 'a', 'a', 'b' , 'B', 'c' ]

pop_with_elements_left(list, "B")

```

Create a function that takes in 3 lists as parameters, and indicates the two that are "closest". That is to say, the two that have the most elements in common.

```python
'''
  This function takes 3 list and shows which two list have the most element in common.

  list1: Input List1 with nElements
  list2: Input List2 with nElements
  list3: Input List3 with nElements
'''
def closest_lists(list1,list2,list3):   
  list1 = set(list1)
  list2 = set(list2)
  list3 = set(list3)
  value1 = list2.intersection(list1)             
  value2 = list3.intersection(list1)
  value3 = list2.intersection(list3)
 
  value_max = max(value1, value2, value3)
  
  if value_max == 0: 
    print("All lists have nothing in common")

  elif value_max == value1: 
    if value_max == value3:
      if value_max == value2:
        print("All lists are equally similar")

      else:
        print("First and second and aswell as second and third list are equally closest")

    elif value_max == value2:
      print("First and second aswll as  first and third list are equally closest")

    else:
      print("First and second list are closest")
      
  elif value_max == value2:
    if value_max == value3:
      print("First and third as well as second and third list are equally close")

    else:
      print("First and third list are closest")
      
  elif value_max == value3:
    print("Second and third list are closest")

list1= ['a', 'b' ,'c']
list2= ['a', 'b', 'd']
list3= ['a', 'c', 'e']

closest_lists(list1,list2,list3)
```
